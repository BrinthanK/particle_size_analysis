```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage
from skimage import measure, color, io
```

```python
# read image for opencv works
img=cv2.imread('C:/Users/brint/Documents/GitHub/particlesize/cropimage.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
plt.show()
```

```python
# creating conversion unit from pixel to mm 
px_mm = 0.55
```

```python
# using otsu's algorithm to threshold the image 
ret, thresh_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret
```

```python
plt.imshow(thresh_img,cmap='gray')
plt.show()
```

```python
# https://docs.opencv.org/4.x/d3/db4/tutorial_py_watershed.html
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh_img,cv2.MORPH_OPEN,kernel, iterations = 1)
```

```python
plt.imshow(opening,cmap='gray')
plt.show()
```

```python
# create sure background 
bg=cv2.dilate(opening,kernel,iterations = 5)
```

```python
plt.imshow(bg,cmap='gray')
plt.show()
```

```python
# distance transfotmation 
dist_transform = cv2.distanceTransform(opening,cv2.DIST_C,3)
```

```python
plt.imshow(dist_transform,cmap='gray')
plt.show()
```

```python
ret2, sure_fg = cv2.threshold(dist_transform,0.10*dist_transform.max(),255,0)
```

```python
print(dist_transform.max())
```

```python
plt.imshow(sure_fg,cmap='gray')
plt.show()
```

```python
sure_fg = np.uint8(sure_fg)

unknown = cv2.subtract(bg,sure_fg)
```

```python
plt.imshow(unknown,cmap='gray')
plt.show()
```

```python
ret3, markers = cv2.connectedComponents(sure_fg)
```

```python
sure_fg
```

```python
markers = markers +1
```

```python
plt.imshow(markers)
plt.show()
```

```python
markers[unknown==255] = 0
```

```python
plt.imshow(markers,cmap='jet')
plt.show()
```

```python
# call back the image in the original colour
img=cv2.imread('C:/Users/brint/Documents/GitHub/particlesize/cropimage.jpg')
markers = cv2.watershed(img,markers)
```

```python
# assign red to the markers with -1 value (i.e., boundary)
img[markers == -1] = [255,0,0]
```

```python
img2 = color.label2rgb(markers, bg_label=0)
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img2,cmap='jet')
plt.show()
```

```python
# measure the regional properties of labelled segments 
# use intital image with the same size for intensity reference 
# use cache to increase the computational efficiency 
img_measure = measure.regionprops(markers,img,cache=True)
```
